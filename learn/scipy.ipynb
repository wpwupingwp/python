## cluster

```python
from scipy import cluster
import numpy as np
from matplotlib import pyplot as plt

rng = np.random.default_rng()
size = 50
x = rng.multivariate_normal([1, 4], [[4, 1], [1, 4]], size=size)
y = rng.multivariate_normal([10, 8], [[10, 2], [2, 10]], size=size)
data = np.concatenate([x, y])
# print('raw', data)
data2 = cluster.vq.whiten(data)
# print('whiten', data2)
codebook, distortion = cluster.vq.kmeans(data2, 2)
# print('k', k)
code, dist = cluster.vq.vq(data, codebook)
print(dist)
centroid_0 = data2[code==0]
centroid_1 = data2[code==1]
plt.scatter(centroid_0[:, 0], centroid_0[:, 1], color='blue', label='type 0')
plt.scatter(centroid_1[:, 0], centroid_1[:, 1], color='green', label='type 1')
plt.scatter(codebook[:, 0], codebook[:, 1], color='red', label='centorid')
plt.show()
plt.legend()
```
