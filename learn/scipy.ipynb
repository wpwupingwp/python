## cluster

```python
from scipy import cluster
import numpy as np
from matplotlib import pyplot as plt

rng = np.random.default_rng()
size = 50
x = rng.multivariate_normal([1, 2], [[4, 1], [1, 4]], size=size)
y = rng.multivariate_normal([10, 8], [[10, 2], [2, 10]], size=size)
data = np.concatenate([x, y])
rng.shuffle(data)
# print('raw', data)
whiten = cluster.vq.whiten(data)
# print('whiten', data2)
codebook, distortion = cluster.vq.kmeans(whiten, 2)
# print('k', k)
code, dist = cluster.vq.vq(whiten, codebook)
centroid_0 = whiten[code==0]
centroid_1 = whiten[code==1]
plt.scatter(data[:, 0], data[:, 1], color='blue', alpha=0.5, label='original value')
plt.scatter(whiten[:, 0], whiten[:, 1], color='green', alpha=0.5, label='whiten value')
plt.scatter(codebook[:, 0], codebook[:, 1], color='red', label='centorid')
plt.legend()
plt.show()
# use kmeans2
centroid, label = cluster.vq.kmeans2(data, 2, minit='points')
c0 = data[label==0]
c1 = data[label==1]
plt.scatter(c0[:, 0], c0[:, 1], color='orange', alpha=0.8, label='cluster 0')
plt.scatter(c1[:, 0], c1[:, 1], color='yellow', alpha=0.8, label='cluster 1')
plt.scatter(centroid[:, 0], centroid[:, 1], color='red', label='centroids')
plt.legend()
plt.show()
```

```python
linkage = cluster.hierarchy.linkage(data, method='centroid')
cluster.hierarchy.dendrogram(linkage)
```

## constants

```python
from scipy import constants

def show(name, value):
    print(name, value)
show('pi', constants.pi)
show('golden ratio', constants.golden_ratio)
show('light speed', constants.speed_of_light)
show('sound speed', constants.speed_of_sound)
show('G', constants.gravitational_constant)
show('alpha partical constants', constants.physical_constants['alpha particle mass'])
show('blue wave', constants.lambda2nu(600))
```
